```python
score=pd.DataFrame({'수학':[90,80,70],'영어':[100,80,95],'과학':[80,95,100]})
score.index=['규리','은제','유지']
score
```

```python
sc=score[:]
```

```python
#inplcae=True해줌으로써 해당 dataframe에 결과 계속 적용되게 함
sc.drop('유지',inplace=True)
sc
```

```python
sc1=score[:]
sc1.drop(['영어','수학'],axis=1,inplace=True)
sc1
```

```python
#loc(라벨index 기준) vs iloc(위치 index 기준)
sc.iloc[0:2]
```

```python
sc.loc['감자']=[90,80,100]
sc
```

```python
#행과 열 transpose
sc=sc.T
```

```python
sc
```

```python
score['총점']=score['수학']+score['영어']+score['과학']
score
```

```python
score['평균']=score['총점']/3
score
```

```python
grade=[]
for avg in score.평균:
    if (avg>=90): grade.append('A')
    elif (avg>=80): grade.append('B')
    elif (avg>=70): grade.append('C')
    elif (avg>=60): grade.append('D')
    else: grade.append('F')
```

```python
score['학점']=grade
```

```python
score
```

```python
def scolarship(row):
    print(row)
    if (row=='A'): return '장학생'
    else: return '비장학생'
#socre dataframe에 성적 column을 추가 -> apply함수를 이용하여 scolarship에 학점 컬럼을 인자로 넘겨줌
score['성적']=score.학점.apply(scolarship)
```

```python
#PassengerId를 기준으로 index지정, usecols사용하여 사용할 column지정
df=pd.read_csv('data/titanic.csv',index_col='PassengerId',usecols=['PassengerId','Survived','Pclass','Name'])
df
```

```python
df.isna()
```

```python
#fillna() Nan값을 해당 값으로 변경시킴
df['Age'].fillna(df['Age'].mean())
```

```python
#생존자 나이 평균
mean1 = df[df['Survived']==1]['Age'].mean()
#사망자 나이 평균
mean0 = df[df['Survived']==0]['Age'].mean()
print(mean1,mean0)
```

```python
#survivied값이 1인 Age값을 indexing -> 해당 Age컬럼의 na값을 평균으로 변경
#survivied값이 0인 Age값을 indexing -> 해당 Age컬럼의 na값을 평균으로 변경
df.loc[df['Survived']==1,'Age'] = df[df['Survived']==1]['Age'].fillna(mean1)
df.loc[df['Survived']==0,'Age'] = df[df['Survived']==1]['Age'].fillna(mean0)
```
