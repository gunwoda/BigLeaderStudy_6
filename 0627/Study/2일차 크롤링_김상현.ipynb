## 홈페이지가 어떻게 만들어져 있는지 잘 알아야 한다.

크롬드라이버에 의해 자동으로 컨트롤된다.
하나하나 클릭해보며 각 성분이 어떤 html로 되어 있는지 확인해본다.

```python
area=driver.find_element(By.XPATH, '//*[@id="SIDO_NM0"]')
```

xpath는 html 요소에서 오른쪽 마우스를 클릭하면 복사할 수 있다. 복사 - xpath

```python
area.send_keys('서울')
```

구 이름을 가져오기

```python
#find_element: 구 이름들을 리스트로 반환

gu_list_raw = driver.find_element(By.XPATH , '//*[@id="SIGUNGU_NM0"]')
```

```python
gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')
print(gu_list)
```

```python
gu_names = [option.get_attribute ('value') for option in gu_list]
gu_names.remove('')
gu_names
```

```python
element = driver.find_element (By.ID, 'SIGUNGU_NM0')
element.send_keys(gu_names[0]) #강남구
```

```python
xpath='''//*[@id= "searRgSelect"]/span'''
element_sel_gu = driver.find_element(By.XPATH , xpath).click()
```

```python
xpath= '''//*[@id= "glopopd_excel"]/span'''
element_get_excel= driver.find_element(By.XPATH , xpath).click()
```

```python


# 반복문을 이용하여 모든 구 엑셀파일 다운로드 진행
for gu in tqdm_notebook(gu_names):
    element = driver.find_element(By.ID, 'SIGUNGU_NM0')#새로운 구 선택
    element.send_keys(gu)
    
    time.sleep(10)  # 데이터 획득 위한 지연 시간

    xpath = '''//*[@id="searRgSelect"]/span''' #조회 버튼
    element_sel_gu = driver.find_element(By.XPATH, xpath).click()
    
    time.sleep(10)

    xpath = '''//*[@id="glopopd_excel"]/span''' #엑셀 저장 버튼
    element_get_excel = driver.find_element(By.XPATH, xpath).click()
    
    time.sleep(10)
```

```python
driver.quit()
```
